### IS Scaling Necessary?

Scaling is performed as a pre-processing step particularly for cluster analysis because we are working in this data set where each variables are in different units i.e. 1000’s ,10000’s and 100’s.

In statistics, standardizing refers to the process of rescaling the values of the variables in your data set so they share a common scale. Often performed as a pre-processing step, particularly for cluster analysis, standardization may be important as we are working with data where each variable has a different unit (in our case 10000’s, 1000’s and 100’s).The reason this importance is particularly high in cluster analysis is because groups are defined based on the distance between points in mathematical space. When we are working with data where each variable means something different, the fields are not directly comparable. In a situation where one field has a much greater range of value than another because the field with the wider range of values likely has greater distances between values, it may end up being the primary driver of what defines clusters.
Standardization helps to make the relative weight of each variable equal by converting each variable to a unit less measure or relative distance.

#### 	APPLY HIERARCHICAL CLUSTERING TO SCALED DATA. IDENTIFY THE NUMBER OF OPTIMUM CLUSTERS USING DENDROGRAM AND BRIEFLY DESCRIBE THEM.

In this dataset we have used agglomerative hierarchical clustering where each customers are considered as a separate clusters. At each step, the two clusters that are most similar are joined into a single new cluster. Once fused, objects are never separated. For the linkage, we have used Ward Linkage method, That is, at each step, the two clusters are fused which result in the least increase in the pooled within-group sum of squares.
A dendrogram is a diagram that shows the hierarchical relationship between objects. It is most commonly created as an output from hierarchical clustering. The main use of a dendrogram is to work out the best way to allocate objects to clusters. The dendrogram below shows the hierarchical clustering of our data set of customers.


```python
from scipy.cluster.hierarchy import dendrogram, linkage #linkage --> merging two clusters
```

```python
link_method = linkage(df_scaled, method = 'ward')
```

```python
dend = dendrogram(link_method)
```

```python
dend = dendrogram(link_method,
                 truncate_mode='lastp',
                 p = 7,
                 )
```

The horizontal axis of the dendrogram represents the distance or dissimilarity between clusters. The vertical axis represents the objects and clusters.
At the start each customer is a separate clusters. According to the dendrogram the customers with most similar patterns in terms of features are merger into a clusters. This process continues until all the customers are segregated into one single clusters representing the whole sample.
These techniques do not let you explicitly set the number of clusters. Instead, you pick a distance value that will yield an appropriate number of clusters so in this case I have decided the number of clusters to be 2 and the reason being as follows:
In the dendrogram above, the height of the dendrogram indicates the order in which the clusters were joined. In this case, the dendrogram shows us that the big difference between clusters is between the clusters in Green versus the clusters in Red.
The green cluster contains the observation in which the customers have most similar patterns in terms of features .We observe that there are 70 customers in the green clusters. The red cluster in other hand though there is distance inside its own clusters the dendrogram suggests that it contains 140 customers in that clusters which suggests that it contains the observations in which the the customers have similar patterns.
The below image shows the means of the 2 clusters from our dataset.


```python
from scipy.cluster.hierarchy import fcluster #forming the clusters
```

```python
# Method 1

clusters = fcluster(link_method, 2, criterion='maxclust')
clusters
```

```python
# Method 2

clusters = fcluster(link_method, 25, criterion='distance')
clusters
```

###  	APPLY K-MEANS CLUSTERING ON SCALED DATA AND DETERMINE OPTIMUM CLUSTERS. APPLY ELBOW CURVE AND SILHOUETTE SCORE.

K-means clustering is one of the simplest and popular unsupervised machine learning algorithms. Kmeans algorithm is an iterative algorithm that tries to partition the dataset into Kpre-defined distinct non-overlapping subgroups i.e. clusters where each data point belongs to only one group. It tries to make the intra-cluster data points as similar as possible while also keeping the clusters as different as possible. It assigns data points to a cluster such that the sum of the squared distance between the data points and the cluster’s centroid (arithmetic mean of all the data points that belong to that cluster) is at the minimum. The less variation we have within clusters, the more homogeneous (similar) the data points are within the same cluster.
We have applied K-Means clustering to the data set using the KMeans function in sklearn package.After applying the Kmeans function , we looked for the within sum of square between each observation and used that to plot the elbow curve to determine the optimum number of clusters.


```python
from sklearn.cluster import KMeans 
```

```python
k_means = KMeans(n_clusters = 2)
```

```python
k_means.fit(df_scaled)
```

```python
k_means.labels_
```

```python
k_means.inertia_ #total wss when k=2
```

```python
wss = []
```

```python
for i in range(1,11):
    KM = KMeans(n_clusters=i)
    KM.fit(df_scaled)
    wss.append(KM.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,11), wss)
```

Elbow method gives us an idea on what a good k number of clusters would be based on the sum of squared distance (SSE) between data points and their assigned clusters’ centroids. We pick k at the spot where SSE starts to flatten out and forming an elbow
Looking at the elbow curve we can see that the curve stats flattening at cluster 3 and hence we select 3 as the optimum clusters.
Silhouette analysis can be used to determine the degree of separation between clusters. the best average silhouette score is around 0.46 for number of clusters equal to 2.However we select 3 as the optimum numner of cluster for this dataset


```python
k_means = KMeans(n_clusters = 3)
k_means.fit(df_scaled)
labels = k_means.labels_
labels
```

```python
k_means = KMeans(n_clusters = 2)
k_means.fit(df_scaled)
labels = k_means.labels_
labels
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
silhouette_score(df_scaled,labels)  # more than 0.5 well distinguished clusters
```

```python
sil_width = silhouette_samples(df_scaled,labels)
sil_width
```

```python
silhouette_samples(df_scaled,labels).min()
```

```python
df["Clus_kmeans"] = labels
df["sil_width"] = sil_width
```
